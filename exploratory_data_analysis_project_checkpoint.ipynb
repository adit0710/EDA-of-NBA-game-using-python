```python
import pandas as pd

# Retrieve HTML table data
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html = pd.read_html(url, header = 0)
df2019 = html[0]

# Data cleaning
raw = df2019.drop(df2019[df2019.Age == 'Age'].index)
raw
```

```python
#data cleaning
```

```python
#data dimension (will give you rows and column)
raw.shape
```

```python
#will display starting entry
raw.head()
```

```python
#check missing value and will add within column
raw.isnull().sum()
```

```python
#replace null with zero
df = raw.fillna(0)
```

```python
#checking again that null is replaced by zero
df.isnull().sum()
```

```python
#will drop rank column as it is not needed
df = df.drop(['Rk'], axis=1)
df
```

```python
#write data in CSV file
```

```python
#index is taken false as we dont need that column
df.to_csv('nba2019.csv', index=False)
```

```python
#reading the csv and storing it in variable
df = pd.read_csv('nba2019.csv')
```

```python
#to display column of data type having number(number includes int and float)
df.select_dtypes(include=['number'])
```

```python
#which player has maximum point
df[df.PTS == df.PTS.max()]
```

```python
#from which team does the player belong to
PlayerMaxPoints = df[df.PTS == df.PTS.max()]
PlayerMaxPoints.Tm
```

```python
#Which position is the player playing as?
PlayerMaxPoints.Pos
```

```python
#Which player scored more than 20 Points (PTS) Per Game?
df[df.PTS > 20]
```

```python
#Which player had the highest 3-Point Field Goals Per Game (3P) ?
df[df['3P'] == df['3P'].max()]
```

```python
#Which player had the highest Assists Per Game (AST) ?
df[df['AST'] == df['AST'].max()]
```

```python
#Which player scored the highest (PTS) in the Los Angeles Lakers?
LAL = df.groupby('Tm').get_group('LAL')
LAL[LAL.PTS == LAL.PTS.max()]
```

```python
df.groupby('Pos').PTS.describe()
```

```python
#Visualization
```

```python
#defining subset which will be used as x axis and y axis in graph 
PTS = df[['Pos','PTS']]

positions = ['C','PF','SF','PG','SG']
PTS = PTS[ PTS['Pos'].isin(positions)  ]

PTS
```

```python
#histogram
PTS['PTS'].hist(by=PTS['Pos'])
```

```python
#with interval of 5
PTS['PTS'].hist(by=PTS['Pos'], layout=(1,5))
```

```python
#scalling the graph with 16:2 ratio
PTS['PTS'].hist(by=PTS['Pos'], layout=(1,5), figsize=(16,2))
```

```python
#representing in boxplot graph
PTS.boxplot(column='PTS', by='Pos')
```
